```python
H = get_SI(6000 * u.watt)
W = get_SI(1500 * u.rpm)
T = H/W
T
```

```python
# parmetros globales
params = {
    "Fn": None,
    "pmax": 2.0e6,
    "theta1": 0.0,
    "theta2": 1.2,
    "r": 0.08,
    "w": 0.05,
    "T": None,
    "Ff": None,
    "mu": 0.35,
    "Fa": None,
    "b": 0.04,
    "a": 0.10,
    "c": 0.08,
    "p": None,
    "thetaMax": np.pi/2,
    "TFn": 500,
    "TFf": 200,
    "Rx": None,
    "Ry": None
}
```

$$
p = p_{max} \, \frac{\sin(\theta_2 - \theta_1)}{\sin(\theta_{max})}
$$


```python
# presion_larga
# objetivos posibles: "p", "pmax", "theta1", "theta2", "thetaMax"
objetivo = "p"
ZE.solve_auto("presion_larga", objetivo, params)
```

$$
T = \frac{w \, r \, b \, p_{max}}{\sin(\theta_{max})}
\left[\frac{1}{2}(\theta_2 - \theta_1) - \frac{1}{4}(\sin(2\theta_2) - \sin(2\theta_1))\right]
$$


```python
# parnormal_larga
# objetivos posibles: "T", "w", "r", "b", "pmax", "theta1", "theta2", "thetaMax"
objetivo = "T"
ZE.solve_auto("parnormal_larga", objetivo, params)
```

$$
T = \frac{w \, r \, b \, p_{max}}{\sin(\theta_{max})}
\left[-r(\cos(\theta_2) - \cos(\theta_1)) - \frac{b}{2}(\sin^2(\theta_2) - \sin^2(\theta_1))\right]
$$


```python
# parfriccion_larga
# objetivos posibles: "T", "w", "r", "b", "pmax", "theta1", "theta2", "thetaMax"
objetivo = "T"
ZE.solve_auto("parfriccion_larga", objetivo, params)
```

$$
F_a = \frac{T_{Fn}}{a} - \frac{T_{Ff}}{a}
$$


```python
# fuerzaaplicacion_energizante_larga
# objetivos posibles: "Fa", "TFn", "TFf", "a"
objetivo = "Fa"
ZE.solve_auto("fuerzaaplicacion_energizante_larga", objetivo, params)
```

$$
F_a = \frac{T_{Fn}}{a} + \frac{T_{Ff}}{a}
$$


```python
# fuerzaaplicacion_desenergizante_larga
# objetivos posibles: "Fa", "TFn", "TFf", "a"
objetivo = "Fa"
ZE.solve_auto("fuerzaaplicacion_desenergizante_larga", objetivo, params)
```

$$
T = \frac{\mu \, w \, r^2 \, p_{max}}{\sin(\theta_{max})}
\left[\cos(\theta_1) - \cos(\theta_2)\right]
$$


```python
# parfrenado_larga
# objetivos posibles: "T", "mu", "w", "r", "pmax", "thetaMax", "theta1", "theta2"
objetivo = "T"
ZE.solve_auto("parfrenado_larga", objetivo, params)
```

$$
R_x = \frac{w \, r \, p_{max}}{2 \, \sin(\theta_{max})}
\left[-\sin^2(\theta_2) + \mu\theta_2 - \frac{1}{2}\sin(2\theta_2)
+ \sin^2(\theta_1) - \mu\theta_1 + \frac{1}{2}\sin(2\theta_1)\right]
$$


```python
# rx_larga
# objetivos posibles: "Rx", "w", "r", "pmax", "mu", "theta1", "theta2", "thetaMax"
objetivo = "Rx"
ZE.solve_auto("rx_larga", objetivo, params)
```

$$
R_y = \frac{w \, r \, p_{max}}{2 \, \sin(\theta_{max})}
\left[-\mu\sin^2(\theta_2) + \theta_2 - \frac{1}{2}\sin(2\theta_2)
+ \mu\sin^2(\theta_1) - \theta_1 + \frac{1}{2}\sin(2\theta_1)\right] - F_a
